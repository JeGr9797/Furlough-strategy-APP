```python

import gurobipy as gp
from gurobipy import GRB
import numpy as np
import time


inicio = time.time()
Iteraciones = 10
obj_values = []
for h in range(Iteraciones):

    Periodos = range(12)
    S = 100
    Escenarios = range(S)
    Productos = range(6)
    Recursos = range(3)
    mu = -0.01
    sigma = 0.07

    #Demanda = np.array([[300,800,400,500,800,600,500,700,800,1100,1200,1300],
                        #[300,800,400,500,800,600,500,700,800,1100,1200,1300],
                        #[500,1000,700,600,1000,800,700,800,1000,1400,1400,1400],
                        #[500,1000,700,600,1000,800,700,800,1000,1400,1400,1400],
                        #[700,1200,1000,700,1200,1000,900,900,1200,1700,1600,1500],
                        #[700,1200,1000,700,1200,1000,900,900,1200,1700,1600,1500]])

    #Escenario 2 ya
    #Demanda = np.array([[3000,4000,3500,3800,4000,3600,3400,3300,3000,3700,3300,3000],
                        #[3000,4000,3500,3800,4000,3600,3400,3300,3000,3700,3300,3000],
                        #[2800,3800,3300,3500,3800,3400,3200,3200,2800,3500,3100,2800],
                        #[2800,3800,3300,3500,3800,3400,3200,3200,2800,3500,3100,2800],
                        #[2600,3600,3100,3200,3600,3200,3000,3100,2600,3300,2900,2600],
                        #[2600,3600,3100,3200,3600,3200,3000,3100,2600,3300,2900,2600]])

    #Escenario 3 ya
    #Demanda = np.array([[300,800,400,500,800,600,500,700,800,3700,3300,3000],
                        #[300,800,400,500,800,600,500,700,800,3700,3300,3000],
                        #[500,1000,700,600,1000,800,700,800,1000,3500,3100,2800],
                        #[500,1000,700,600,1000,800,700,800,1000,3500,3100,2800],
                        #[700,1200,1000,700,1200,1000,900,900,1200,3300,2900,2600],
                        #[700,1200,1000,700,1200,1000,900,900,1200,3300,2900,2600]])

    #Escenario 4 ya
    #Demanda = np.array([[3000,4000,3500,500,800,600,500,700,800,1100,1200,1300],
                       #[3000,4000,3500,500,800,600,500,700,800,1100,1200,1300],
                        #[2800,3800,3300,600,1000,800,700,800,1000,1400,1400,1400],
                       #[2800,3800,3300,600,1000,800,700,800,1000,1400,1400,1400],
                        #[2600,3600,3100,700,1200,1000,900,900,1200,1700,1600,1500],
                        #[2600,3600,3100,700,1200,1000,900,900,1200,1700,1600,1500]])

    #Escenario 5 ya
    #Demanda = np.array([[300,800,400,3800,4000,3600,3400,3300,3000,3700,3300,3000],
                        #[300,800,400,3800,4000,3600,3400,3300,3000,3700,3300,3000],
                        #[500,1000,700,3500,3800,3400,3200,3200,2800,3500,3100,2800],
                        #[500,1000,700,3500,3800,3400,3200,3200,2800,3500,3100,2800],
                        #[700,1200,1000,3200,3600,3200,3000,3100,2600,3300,2900,2600],
                        #[700,1200,1000,3200,3600,3200,3000,3100,2600,3300,2900,2600]])

    #Escenario 6 ya
    Demanda = np.array([[3000,4000,3500,500,800,600,500,700,800,3700,3300,3000],
                        [3000,4000,3500,500,800,600,500,700,800,3700,3300,3000],
                        [2800,3800,3300,600,1000,800,700,800,1000,3500,3100,2800],
                        [2800,3800,3300,600,1000,800,700,800,1000,3500,3100,2800],
                        [2600,3600,3100,700,1200,1000,900,900,1200,3300,2900,2600],
                        [2600,3600,3100,700,1200,1000,900,900,1200,3300,2900,2600]])


    
    P_t = {}
    P_t_1 = {}
    
    for i in Productos:
        for t in Periodos:
            for r in Recursos: 
                for s in Escenarios:
                    if t == 0 and r == 0 and i == 0:
                        P_t[i,t,r,s] = 1
                        P_t_1[i,t,r,s] = 1
                    if t == 0 and r == 0 and i == 1:
                        P_t[i,t,r,s] = 1
                        P_t_1[i,t,r,s] = 1
                    if t == 0 and r == 0 and i == 2:
                        P_t[i,t,r,s] = 1.5
                        P_t_1[i,t,r,s] = 1.5
                    if t == 0 and r == 0 and i == 3:
                        P_t[i,t,r,s] = 1.5
                        P_t_1[i,t,r,s] = 1.5
                    if t == 0 and r == 0 and i == 4:
                        P_t[i,t,r,s] = 2
                        P_t_1[i,t,r,s] = 2
                    if t == 0 and r == 0 and i == 5:
                        P_t[i,t,r,s] = 2
                        P_t_1[i,t,r,s] = 2
                    if t == 0 and r == 1 and i == 0:
                        P_t[i,t,r,s] = 1.5
                        P_t_1[i,t,r,s] = 1.5
                    if t == 0 and r == 1 and i == 1:
                        P_t[i,t,r,s] = 1.5
                        P_t_1[i,t,r,s] = 1.5
                    if t == 0 and r == 1 and i == 2:
                        P_t[i,t,r,s] = 2
                        P_t_1[i,t,r,s] = 2
                    if t == 0 and r == 1 and i == 3:
                        P_t[i,t,r,s] = 2
                        P_t_1[i,t,r,s] = 2
                    if t == 0 and r == 1 and i == 4:
                        P_t[i,t,r,s] = 2.5
                        P_t_1[i,t,r,s] = 2.5
                    if t == 0 and r == 1 and i == 5:
                        P_t[i,t,r,s] = 2.5
                        P_t_1[i,t,r,s] = 2.5
                    if t == 0 and r == 2 and i == 0:
                        P_t[i,t,r,s] = 0.5
                        P_t_1[i,t,r,s] = 0.5
                    if t == 0 and r == 2 and i == 1:
                        P_t[i,t,r,s] = 0.5
                        P_t_1[i,t,r,s] = 0.5
                    if t == 0 and r == 2 and i == 2:
                        P_t[i,t,r,s] = 0.5
                        P_t_1[i,t,r,s] = 0.5
                    if t == 0 and r == 2 and i == 3:
                        P_t[i,t,r,s] = 0.5
                        P_t_1[i,t,r,s] = 0.5
                    if t == 0 and r == 2 and i == 4:
                        P_t[i,t,r,s] = 0.5
                        P_t_1[i,t,r,s] = 0.5
                    if t == 0 and r == 2 and i == 5:
                        P_t[i,t,r,s] = 0.5
                        P_t_1[i,t,r,s] = 0.5
                    if t > 0:
                        P_t[i,t,r,s] = P_t[i,t-1,r,s]*np.random.lognormal(mu, sigma)
                        P_t_1[i,t,r,s] = P_t_1[i,t-1,r,s]*np.random.lognormal(mu, sigma)
                        
    P_t_bar = np.array([[1,1,1.5,1.5,2,2],
                        [1.5,1.5,2,2,2.5,2.5],
                        [0.5,0.5,0.5,0.5,0.5,0.5]])
    
    p = np.array([1.0/S]*S)
    V = np.array([300,350,450,500,600,650])
    C_1 = np.array([50,60,80,90,110,120])
    C_2 = np.array([15,20,30,35,45,50])
    C_3 = np.array([140,150,170,180,200,210])
    C_4 = np.array([30,35,45,50,60,65])
    C_5 = np.array([150,160,180,190,210,220])
    C_6 = np.array([6.25,5.625,5])
    C_7 = np.array([100,90,80])
    C_8 = np.array([250,250,250])
    

    N_time = 40
    O_time = 5
    Min_cap = 0.85 
    M = 10000
    #Sensibilidad
    Budget_workers = [5,5,5]
    Lt = 1


    #Declarar e iniciar modelo optimizacion
    #Declarar e iniciar modelo optimizacion
    m = gp.Model("Lot_sizing_paper")

    #Variables de decision
    #Primer estado
    w = m.addVars(Periodos, Recursos, Escenarios, vtype=GRB.INTEGER,name="w")
    h = m.addVars(Periodos, Recursos, Escenarios, vtype=GRB.INTEGER,name="h")
    hw_1 = m.addVars(Periodos, Recursos, Escenarios, name="horas_normal")
    hw_2 = m.addVars(Periodos, Recursos, Escenarios, name="horas_extra")
    etha = m.addVars(Periodos, Recursos, Escenarios, vtype=GRB.BINARY, name="furlough")

    #Recurso o segundo estado
    Inv = m.addVars(Productos, Periodos, Escenarios, vtype=GRB.INTEGER, lb = 0, name="Inventario")
    x = m.addVars(Productos, Periodos, Escenarios, lb = 0, name="Produccion_normal")
    o = m.addVars(Productos, Periodos, Escenarios,vtype=GRB.INTEGER, lb = 0,name="Produccion_extra")
    b = m.addVars(Productos, Periodos, Escenarios,vtype=GRB.INTEGER, lb = 0, name="Retrasos")
    o_t = m.addVars(Productos, Periodos, Escenarios, vtype=GRB.BINARY, lb = 0, name="o_t")
   
    #Funcion objetivo
    #Sentido modelo
    m.ModelSense = GRB.MAXIMIZE
    #Costos independientes
    Costo_total_produccion = gp.quicksum(C_1[i]*x[i,t,s]*p[s] 
                                        for i in Productos for t in Periodos for s in Escenarios )
    Costo_total_inventario = gp.quicksum(C_2[i]*Inv[i,t,s]*p[s] 
                                        for i in Productos for t in Periodos for s in Escenarios )
    Costo_total_produccion_extra = gp.quicksum(C_3[i]*o[i,t,s]*p[s] 
                                        for i in Productos for t in Periodos for s in Escenarios )
    Costo_total_retrasos = gp.quicksum(C_4[i]*b[i,t,s]*p[s] 
                                        for i in Productos for t in Periodos for s in Escenarios )
    Costo_total_incurrir_extra = gp.quicksum(C_5[i]*o_t[i,t,s]*p[s] 
                                        for i in Productos for t in Periodos for s in Escenarios )
    Costo_total_recurso_tNormal = gp.quicksum(C_6[r]*hw_1[t,r,s]*w[t,r,s]*p[s]  
                                        for t in Periodos for r in Recursos for s in Escenarios )
    Costo_total_recurso_tExtra = gp.quicksum(C_6[r]*hw_2[t,r,s]*w[t,r,s]*p[s]  
                                        for t in Periodos for r in Recursos for s in Escenarios )
    Costo_total_recurso_contratar = gp.quicksum(C_7[r]*h[t,r,s]*p[s]  
                                        for t in Periodos for r in Recursos for s in Escenarios )
    Costo_total_politica_hrs = gp.quicksum(C_8[r]*etha[t,r,s]*w[t,r,s]*p[s]  
                                        for t in Periodos for r in Recursos for s in Escenarios )
    Utilidades_1 = gp.quicksum(V[i]*x[i,t,s]*p[s]
                            for i in Productos for t in Periodos for s in Escenarios)
    Utilidades_2 = gp.quicksum(V[i]*o[i,t,s]*p[s]
                            for i in Productos for t in Periodos for s in Escenarios)
    #Costo total
    Utilidad_total = (Utilidades_1 + Utilidades_2 -
                      Costo_total_produccion - 
                      Costo_total_inventario -
                      Costo_total_produccion_extra -
                      Costo_total_retrasos - 
                      Costo_total_incurrir_extra - 
                      Costo_total_recurso_tNormal -
                      Costo_total_recurso_tExtra -
                      Costo_total_recurso_contratar -
                      Costo_total_politica_hrs)

    #definir modelo
    m.setObjective(Utilidad_total)

    #Ecuaciones balanceo de recursos
    #Existe fuerza laboral
    m.addConstrs((w[t,r,s] == 6)
                for t in Periodos if t == 0 for r in Recursos for s in Escenarios)

    m.addConstrs((w[t,r,s] == w[t-1,r,s])
                for t in Periodos if t >= 1 and t <= Lt for r in Recursos for s in Escenarios)
    
    m.addConstrs((w[t,r,s] == w[t-1,r,s] + h[t-Lt,r,s])
                for t in Periodos if t >= 1 and t > Lt for r in Recursos for s in Escenarios)

    #Ecuaciones balanceo de produccion
    m.addConstrs((x[i,t,s] + o[i,t,s] == Demanda[i,t] + Inv[i,t,s] - b[i,t,s]) 
                for i in Productos for t in Periodos if t == 0 for s in Escenarios )

    m.addConstrs((Inv[i,t-1,s] + x[i,t,s] + o[i,t,s] == Demanda[i,t] + Inv[i,t,s] + b[i,t-1,s] - b[i,t,s]) 
                    for i in Productos for t in Periodos if t>=1 for s in Escenarios )
    
    #Inventario final sea cero
    m.addConstrs((Inv[i,t,s] == 0) for i in Productos for t in Periodos if t==11 for s in Escenarios )


    #Restriccion capacidad
    m.addConstrs((gp.quicksum(x[i,t,s]*P_t[i,t,r,s]/60 for i in Productos) <= w[t,r,s]*hw_1[t,r,s] ) 
                    for t in Periodos for r in Recursos for s in Escenarios )

    m.addConstrs((gp.quicksum(o[i,t,s]*P_t_1[i,t,r,s]/60 for i in Productos) <= w[t,r,s]*hw_2[t,r,s] ) 
                    for t in Periodos for r in Recursos for s in Escenarios )
    
    #activacion restricciones logicas
    m.addConstrs((o[i,t,s] <= M*o_t[i,t,s])
                    for i in Productos for t in Periodos for s in Escenarios )
    
    
    #Maximos horas
    m.addConstrs((hw_1[t,r,s] <= N_time)
            for t in Periodos for r in Recursos for s in Escenarios )
    m.addConstrs(hw_2[t,r,s] <= O_time*(1-etha[t,r,s]) 
            for t in Periodos for r in Recursos for s in Escenarios )
    
    #logica activar reduccion horas
    m.addConstrs((Min_cap*w[t,r,s]*N_time - gp.quicksum(Demanda[i,t]*P_t_bar[r,i]/60 for i in Productos)
                  <= M*etha[t,r,s])
            for t in Periodos for r in Recursos for s in Escenarios )
    
    m.addConstrs((N_time - hw_1[t,r,s] <= N_time*etha[t,r,s])
            for t in Periodos for r in Recursos for s in Escenarios )
    
    #Si hay reduccion horas, no se puede contratar
    m.addConstrs((h[t,r,s] <= Budget_workers[r]*(1-etha[t,r,s]))
            for t in Periodos for r in Recursos for s in Escenarios)
                 
    #Maximo trabajadores 
    m.addConstrs((gp.quicksum(h[t,r,s] for t in Periodos) <= Budget_workers[r])
                for r in Recursos for s in Escenarios)
    
    #restricciones de no anticipacion
    m.addConstrs((w[t,r,s] == w[t,r,s-1]) 
                    for t in Periodos for r in Recursos for s in Escenarios if s >= 1)

    m.addConstrs((h[t,r,s] == h[t,r,s-1]) 
                    for t in Periodos for r in Recursos for s in Escenarios if s >= 1)

    m.addConstrs((hw_1[t,r,s] == hw_1[t,r,s-1]) 
                    for t in Periodos for r in Recursos for s in Escenarios if s >= 1)

    m.addConstrs((hw_2[t,r,s] == hw_2[t,r,s-1]) 
                    for t in Periodos for r in Recursos for s in Escenarios if s >= 1)

    m.addConstrs((etha[t,r,s] == etha[t,r,s-1]) 
                    for t in Periodos for r in Recursos for s in Escenarios if s >= 1)  


    m.setParam('MIPGap', 0.01)
    m.params.NonConvex = 2
    m.write("checar.lp")
    #optimizar
    m.optimize()
    #Guardar soluciones objetivo y actualizar
    obj_sample = m.objVal
    obj_values.append(obj_sample)
    
    tot_dem_FG = {}
    tot_sup_FG = {}
    tot_back_FG ={}
    
    for i in Productos:
        tot_dem_FG[i] = 0
        tot_sup_FG[i] = 0
        tot_back_FG[i] = 0
        for t in Periodos:
            tot_dem_FG[i] = tot_dem_FG[i] + Demanda[i,t]
            tot_sup_FG[i] += sum(x[i,t,s].x*p[s] + o[i,t,s].x*p[s] for s in Escenarios)
            tot_back_FG[i] = tot_dem_FG[i] - tot_sup_FG[i]
    
    Porcentaje_demanda_cumplido = {}
    for i in Productos:
        print(f"*----------------------------*")
        print(f"producto {i}:")
        Porcentaje_demanda_cumplido[i] = 0
        Porcentaje_demanda_cumplido[i] = tot_sup_FG[i]/tot_dem_FG[i]
        print(f"porcentaje demanda utilizada es: {100*Porcentaje_demanda_cumplido[i]} %")
    print(f"*----------------------------*")
    print(f"-------------------------------------------------------------------------------------------------------")
    
    consumed_capacity = {}
    available_capacity = {}
    average_consumed_capacity = {}
    average_available_capacity = {}
    average_resource_consumption = {}
    for t in Periodos:
        for r in Recursos:
            for s in Escenarios:
                key = (t, r, s)
                consumed_capacity[key] = 0.0
                consumed_capacity[t,r,s] += sum(x[i,t,s].x*P_t[i,t,r,s]*p[s]/60 for i in Productos for s in Escenarios)
                available_capacity[t,r,s] = w[t,r,s].x*hw_1[t,r,s].x
    for r in Recursos:
        print(f"Trabajo {r}:")
        key_1 = (r)
        average_consumed_capacity[key_1] = 0.0
        average_consumed_capacity[r] += sum(x[i,t,s].x*P_t[i,t,r,s]*p[s]/60 for i in Productos for t in Periodos for s in Escenarios)
        average_available_capacity[r] = sum(w[t,r,s].x*hw_1[t,r,s].x for t in Periodos)
        average_resource_consumption = round(100*average_consumed_capacity[r]/average_available_capacity[r],2)
        print(f"el consumo por recurso fue de:",average_resource_consumption,"%")
                
    print(f"-------------------------------------------------------------------------------------------------------")

print(f"Valor esperado LB es: {np.mean(obj_values)}")
print(f"Desviación estandar LB es: {np.std(obj_values)}")
fin = time.time()
t_total = fin-inicio 
print("Tiempo total:",t_total)


    

```
